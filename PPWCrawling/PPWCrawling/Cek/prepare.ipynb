```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import pandas as pd

df = pd.read_csv('t.csv', delimiter=',')
```

```python
import numpy as np
contents = df['Contents']
print(np.array(contents))
```

```python
word_tokens = word_tokenize(contents[1])
print(word_tokens)
```

```python
stop_words = set(stopwords.words('indonesian'))

word_tokens_no_stopwords = [w for w in word_tokens if not w in stop_words]
print(word_tokens_no_stopwords)
```

```python
# initializing special characters
special_char = '@_!#$%^&*()<>?/\|}{~:;.[],-1234567890'

# using join() + generator to remove special characters
out_list = [''.join(x for x in string if not x in special_char) for string in word_tokens_no_stopwords]

# print list without special characters
print('List after removal of special characters:', out_list)
```

```python
out_list.remove('')
print(out_list)

freq_kata_1 = nltk.FreqDist(word_tokens)

freq_kata_2 = nltk.FreqDist(out_list)
# print(freq_kata_1.most_common(20))

print(freq_kata_2.most_common(len(out_list)))
```
