# Memasukan CSV

```python
#Rizki Ashuri Pratama
#190411100151
import pandas as pd
import numpy as np

#Disesuaikan Letak file
#Membuat Dataframe
dataURL = pd.read_csv('C:\Data Kuliah\Kuliah\Semester 6\Pencarian dan Penambangan Web\PPWCrawling\PPWCrawling\Berita\LinkBerita.csv')
dataKonten = pd.read_csv('C:\Data Kuliah\Kuliah\Semester 6\Pencarian dan Penambangan Web\PPWCrawling\PPWCrawling\Berita\KontenBerita.csv')
dataBerita = pd.concat([dataURL, dataKonten], axis=1)

#Menampilkan dataBerita
dataBerita
```

# Case Folding (Membuat Non Kapital)

```python
#Case Folding untuk membuat huruf menjadi non kapital
dataBerita['Konten'] = dataBerita['Konten'].str.lower()

#Menampilkan Hasil Casefolding pada kolom Konten
dataBerita['Konten']
```

# Tokenizing (Memecah string)

```python
import string
from nltk.tokenize import word_tokenize
import re

def remove_tweet_special(text):
    # remove tab, new line, ans back slice
    text = text.replace('\\t'," ").replace('\\n'," ").replace('\\u'," ").replace('\\',"")
    # remove non ASCII (emoticon, chinese word, .etc)
    text = text.encode('ascii', 'replace').decode('ascii')
    # remove mention, link, hashtag
    text = ' '.join(re.sub("([@#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ", text).split())
    # remove incomplete URL
    return text.replace("http://", " ").replace("https://", " ")
                
dataBerita['Konten'] = dataBerita['Konten'].apply(remove_tweet_special)

#remove number
def remove_number(text):
    return  re.sub(r"\d+", "", text)

dataBerita['Konten'] = dataBerita['Konten'].apply(remove_number)

#remove punctuation
def remove_punctuation(text):
    return text.translate(str.maketrans("","",string.punctuation))

dataBerita['Konten'] = dataBerita['Konten'].apply(remove_punctuation)

#remove whitespace leading & trailing
def remove_whitespace_LT(text):
    return text.strip()

dataBerita['Konten'] = dataBerita['Konten'].apply(remove_whitespace_LT)

#remove multiple whitespace into single whitespace
def remove_whitespace_multiple(text):
    return re.sub('\s+',' ',text)

dataBerita['Konten'] = dataBerita['Konten'].apply(remove_whitespace_multiple)

# remove single char
def remove_singl_char(text):
    return re.sub(r"\b[a-zA-Z]\b", "", text)

dataBerita['Konten'] = dataBerita['Konten'].apply(remove_singl_char)

# NLTK word tokenize 
def word_tokenize_wrapper(text):
    return word_tokenize(text)

dataBerita['Konten'] = dataBerita['Konten'].apply(word_tokenize_wrapper)
dataBerita['Konten']
```

# Stopword (Filter Kata yang penting)

```python
from nltk.corpus import stopwords
#Mendapatkan stopword indonesia
list_stopwords = stopwords.words('indonesian')

#Menghapus Stopword dari list token
def stopwords_removal(words):
    return [word for word in words if word not in list_stopwords]
dataBerita['Konten'] = dataBerita['Konten'].apply(stopwords_removal)

dataBerita['Konten']
```

# Stemming (Mencari Kata Dasar tanpa Imbuhan)

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter

# Membuat Stmmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in dataBerita['Konten']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '
            
print(len(term_dict))
print("------------------------")

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)
    print(term,":" ,term_dict[term])
    
print(term_dict)
print("------------------------")

#Proses Stemming
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

dataBerita['Konten'] = dataBerita['Konten'].swifter.apply(get_stemmed_term)
dataBerita['Konten']
```

```python
#simpan jadi csv
#Nama file : TextProcessing.csv
dataBerita.to_csv('TextPreprocessing.csv')
```
